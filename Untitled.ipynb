```python
import re 
import cv2
def sorted_nicely( l ): 
    """ Sort the given iterable in the way that humans expect.""" 
    convert = lambda text: int(text) if text.isdigit() else text 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)
import glob
import numpy as np
```

```python
#segmentation mask ra image ko list ho 
real_images = sorted_nicely(glob.glob('/mnt/Enterprise/prashant/UACANET/data/TestDataset/CVC-300/images/*.png'))
mask_images = sorted_nicely(glob.glob('/mnt/Enterprise/prashant/UACANET/snapshots/UACANet-L/CVC-300/*.png'))
```

```python
output_video = 'output_video.mp4'

frame_size = (700, 300)
fps = 30.0
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
# video_writer = cv2.VideoWriter(output_video, fourcc, fps, frame_size)
video_writer = cv2.VideoWriter(output_video,fourcc,fps,frame_size)

print(len(real_images), len(mask_images))

for idx in range(len(real_images)):
    real_image = cv2.imread(real_images[idx])
    mask = cv2.imread(mask_images[idx], cv2.IMREAD_GRAYSCALE)
    
    color = np.array([0,255,0], dtype='uint8')
    masked_img = np.where(mask[...,None], color, real_image)
    
    frame = cv2.addWeighted(real_image, 0.1, masked_img, 0.9,0)
    frame = cv2.hconcat([real_image, masked_img])
    
    frame = cv2.resize(frame, frame_size)
    
    # Ensure the frame has 3 channels (RGB)
    if frame.shape[2] != 3:
        frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2RGB)
    
    # Write the frame to the video
    video_writer.write(frame)
    
video_writer.release()
```
